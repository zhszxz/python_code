```python
import pandas

# 构造DataFrame:类似于数据表(学员成绩信息)
df = pandas.DataFrame([
    {"姓名": "小王", "语文": 90, "数学": 80, "英语": 70},
    {"姓名": "小李", "语文": 80, "数学": 90, "英语": 80},
    {"姓名": "小张", "语文": 70, "数学": 80, "英语": 90},
    {"姓名": "小赵", "语文": 60, "数学": 70, "英语": 80},
    {"姓名": "小孙", "语文": 80, "数学": 90, "英语": 70},
    {"姓名": "小李", "语文": 80, "数学": 80, "英语": 90},
    {"姓名": "小王2", "语文": 90, "数学": 80, "英语": 80},
    {"姓名": "小王3", "语文": 65, "数学": 55, "英语": 99},
    {"姓名": "小王4", "语文": 91, "数学": 78, "英语": 82}
])

# Series:相当于表格中的一列
print(type(df['语文']))

print(f"语文最高分: {df['语文'].max()}, 最低分: {df['语文'].min()}, 平均分: {df['语文'].mean():.2f}")
print(f"数学最高分: {df['数学'].max()}, 最低分: {df['数学'].min()}, 平均分: {df['数学'].mean():.2f}")
print(f"英语最高分: {df['英语'].max()}, 最低分: {df['英语'].min()}, 平均分: {df['英语'].mean():.2f}")
```
