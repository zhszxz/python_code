```python
import pandas as pd

pd.set_option('display.min_rows', 30)  # 设置显示行数

df = pd.read_csv('data/sales.csv')

# 处理缺失值
# 查看缺失值
df.isnull()

# 删除缺失值
df.dropna()  #删除行
df.dropna(axis=1)  #删除列

# 填充缺失值
df.fillna('未知')  #填充指定值
df.ffill()  #填充上一行数据
df.bfill()  #填充下一行数据
```

```python
#处理重复值
#查看重复值
df.duplicated()  #默认两行完全相同才算重复
df.duplicated(subset=['订单号'])  #指定判重列

# 删除重复值
df.drop_duplicates(subset=['订单号'], keep='first')
```

```python
# 处理异常值
# 查看异常值
df[df['单价'] < 0]

# 删除异常值
df.drop(df[df['单价'] < 0].index)

# 修复异常值
df['单价']=df['单价'].abs()
df
```

```python
# 处理格式
df['订单日期']=df['订单日期'].str.replace('/','-')
df
```
