## 1.series的创建

```python
import pandas as pd

# 创建series,索引默认从0开始
s = pd.Series([10, 2, 3, 4, 5])

# 自定义索引
s = pd.Series([10, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'])

# 自定义名称
s = pd.Series([10, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'], name="月份")

# 通过字典创建series,使用key作为索引
s = pd.Series({"a": 1, "b": 2, "c": 3, "d": 4, "e": 5})

# 通过series创建series
s2 = pd.Series(s, index=["a", "c"])
print(s2)
```

## 2.series的属性

```python
# 索引
s.index
# 值
s.values
# 形状
s.shape
# 维度
s.ndim
# 元素个数
s.size
s.name = 'test'
# 数据类型
s.dtype
# 名称
s.name
# 通过显式索引取值,支持单值,多值,切片
s.loc['a']
# 通过隐式索引取值,支持单值,多值,切片
s.iloc[1]
# 通过显式索引取单值
s.at['a']
# 通过隐式索引取单值
s.iat[0]
```

```python
# 访问数据
# 通过自定义索引
print(s['c'])
# 通过布尔索引
print(s[s < 4])
# 修改数据
s['e'] = 6
print(s)
```

```python
'''
创建一个包含10名学生数学成绩的Series，成绩范围在50-100之间。
计算平均分、最高分、最低分，并找出高于平均分的学生人数。
'''
import pandas as pd
import numpy as np

np.random.seed(42)
values = np.random.randint(50, 101, 10)
indexes = []
for i in range(1, 11):
    indexes.append('学生%d' % i)
s = pd.Series(values, indexes)
print(s)
print('平均分:', s.mean())
print('最高分:', s.max())
print('最低分:', s.min())
print('高于平均分的学生人数:', s[s > s.mean()].count())
```

```python
'''
温度数据统计
给定某城市一周每天的最高温度Series，完成以下任务：
找出温度超过30度的天数
计算平均温度
将温度从高到低排序
找出温度变化最大的两天
'''
import pandas as pd

temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33],
                         index=['周一', '周二', '周三', '周四', '周五', '周六', '周日'])
print('温度超过30度的天数:', temperatures[temperatures > 30].count())
print('平均温度:', temperatures.mean())
print('温度从高到低排序:', temperatures.sort_values(ascending=False))
print('温度变化最大的两天:', temperatures.diff().idxmax())
```

```python
'''
股票价格分析
给定某股票连续10个交易日的收盘价：
计算每日收益率（当日收盘价/前日收盘价 - 1）
找出收益率最高和最低的日期
计算波动率（收益率的标准差）

prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2], index=pd.date_range('2023-01-01', periods=10))
'''
```

```python
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
                   index=pd.date_range('2023-01-01', periods=10))
s = prices.pct_change()
print(s)
print('收益率最高的日期:', s.idxmax())
print('收益率最低的日期:', s.idxmin())
print('波动率:', s.std())
```

```python
'''
销售数据分析
某产品过去12个月的销售量：
计算季度平均销量（每3个月为一个季度）
找出销量最高的月份
计算月环比增长率
找出连续增长超过2个月的月份
'''
```

```python
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220],
                  index=pd.date_range('2022-01-01', periods=12, freq='MS'))
print(sales)
```

```python
# 季度平均销量
print(sales.resample('QE').mean())
# 销量最高的月份
print(sales.idxmax())
# 月环比增长率
print(sales.pct_change())
# 连续增长超过2个月的月份
b = sales.pct_change() > 0
b[b.rolling(3).sum() == 3].keys().tolist()
```
